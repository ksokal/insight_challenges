# Introduction

Customer engagement is an important metric for understanding the ongoing customer relationships to our product and services. Therefore, the identification in an apparent drop in customer engagement, significant by eye and lasting over a week, is a red flag warranting immediate attention and investigation.

In order to quickly identify the cause of the drop in engagement, I follow the basic path along which we measure customer engagement. Using the customer flow will pinpoint where the most major problem occurred, and further investigations can then proceed at the identified point.

Analysis path:
- explore Yammer interactions (logins, views, clicks, etc.)
- path to Yammer is through emails, then investigate email events

# Take away

After recreating the reported drop in engagement plot, I then examined the types of customer engagement events to identify at which point there was a problem. It is clear that the drop had already occured before the customers were interacting on the Yammer platform. Investigating the customer email interactions, which lead them to the platform, indicate the source of the problem. Customer's opened email at the same rate, but the click rate dropped significantly starting in July.

Therefore, it is apparent that there is an issue with the ability to click on our Yammer link given in emails. The cause, whether a typo, change in appearance/layout, broken link, or down server, needs to be followed up promptly.

```python
def get_engaged_users(d):
    print(d)
    print(type(d))
    try:
        #if d < dailysum.date(len(dailysum)-1):
        week=d+pd.DateOffset(weeks=1)
        print(d, week)
        print('OK NOW HERE')
        weekdata=events[(events["date"] >= d)]
        weekdata=weekdata[(weekdata["date"] <= week)]
        print(len(weekdata))
        #print(weekdata.head())
        weekusers=weekdata[weekdata["event_type"]=="engagement"]["user_id"].nunique()
        #print(weekusers)
        #dailysum["engaged_weekly_users"]=weekusers
        print('ave week users', weekusers)
        return weekusers
    except Exception as e:
        print(e)
```

```python
dailysum["engaged_weekly_users"]=dailysum["date"].apply(lambda x: get_engaged_users(x))
```

```python
def get_engaged_users_specifics(d,col,choice):
    print(d)
    print(type(d))
    try:
        #if d < dailysum.date(len(dailysum)-1):
        week=d+pd.DateOffset(weeks=1)
        print(d, week)
        print('OK NOW HERE')
        weekdata=events[(events["date"] >= d)]
        weekdata=weekdata[(weekdata["date"] <= week)]
        print(len(weekdata))
        #print(weekdata.head())
        weekusers=weekdata[weekdata["event_type"]=="engagement"]
        weekusers=weekusers[weekusers[col] == choice]
        weekusers=weekusers["user_id"].nunique()
        #print(weekusers)
        #dailysum["engaged_weekly_users"]=weekusers
        print('ave week users', weekusers)
        return weekusers
    except Exception as e:
        print(e)
```
