Briefing:

Background:

Basically in 2015, the 'core' countries (i.e. northern and southern European states, who fund then event) started to complain that eastern Europe was acting as a voting block and skewing results, not giving any other countries a change to win.

From 2016 to Present: a hybrid of panel judges and popular vote.

# First thoughts

We should naively assume that any country has the same chance to win every year (thus prob = 1/50) and that previous years are independent of each other (ignoring sociology here).

(1)
The null hypothesis = every country has a fair chance
The alt hypothesis = there was a voting block skewing results

(2) 
The null hypothesis = the voting system doesn't matter
The alt hypothesis = the voting system change in 2015 changed the results. If alt(1) was true, it successfully fixed the skew.

approaches:
- who is winning more than that?
- who does and doesn't vote?
- what do the voting distributions look like for winners compared to the rest? over the years?
- who is voting for the winners vs who is voting randomly?

furthermore:
- can we find clusters?
- can we do network analysis?

```python
winners=pd.DataFrame(orig_data["Year"].unique())
winners.rename(columns={0:"Year"}, inplace=True)
winners["Winner"]=None
winners.tail()
```

#want a table with the 'to winner' only:
year, winner, from country, points
(so year and winner are repeated i guess)

```python
winningvotes=pd.DataFrame()

for year in finals["Year"].unique():
    #finding the winner each year
    winner=np.argmax(finals[finals["Year"]==year].groupby('To country')["Points"].sum())
    
    #update the winners table with the winner
    winners.loc[winners["Year"] == year,"Winner"] = winner
    
    #pull out yearly finals data
    year_data=finals[finals["Year"]==year]
    year_data=year_data[year_data["To country"] == winner]
    winningvotes=winningvotes.append(year_data)
```

statistical test if the high/low lines are different.

```python
sums=pd.DataFrame(winning_pivot_j.count().sort_values())
sums.rename(columns={0:"Votes to winner"}, inplace=True)
```

```python
#sums=sums.reset_index()
sums["N wins"]=sums["From country"].apply(lambda x: winners["Year"][winners["Winner"]==x].count())
```

```python
sums.plot.scatter(x="Votes to winner", y="N wins")
```

```python
#Lets do a T test!
# null = they are the same
# alt = the low and high distributions are different

#splitting by eye for now, although really you would want to run a clustering algo or something to define the break
low=sums[sums["Votes to winner"] < 30]
high=sums[sums["Votes to winner"] > 30]
```

```python
import scipy.stats
```

```python
scipy.stats.ttest_ind(low["N wins"],high["N wins"], axis=0)
```
